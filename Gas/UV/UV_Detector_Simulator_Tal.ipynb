# Filters- Defining filters 

```python

def gaussian(x, mu, sig):
    # Parameters:
    # x: Array of x values.
    # mu: Mean of the Gaussian filter.
    # sig: Standard deviation of the Gaussian filter.

    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

#x=x.astype(np.float) # Optional

```

```python
# Filters 220,254,275 and their preferences:

filter1c=np.ones(end1+1)
filter2c=np.ones(end1+1)
WL=WL.astype(np.float64) # Ensure wavelength data is in float64 format
#WL=WL.float() #optional

# Gaussian factor calculation
GF = np.sqrt(-2 * np.log(0.5))  

# Filter parameters
mu_1, sig_1, PS1 = 220, 5 / GF, 6 / 18
mu_2, sig_2, PS2 = 254, 5 / GF, 18 / 18
mu_3, sig_3, PS3 = 275, 5 / GF, 10 / 18
mu_4, sig_4, PS4 = 230, 10 / GF, 6 / 18

# Creating filters using the Gaussian function
filter1=PS1*gaussian(WL, mu_1, sig_1)
filter2=PS2*gaussian(WL, mu_2, sig_2)
filter3=PS3*gaussian(WL, mu_3, sig_3)
filter4=PS4*gaussian(WL, mu_4, sig_4)
```
