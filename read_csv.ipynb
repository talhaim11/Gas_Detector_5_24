```python
def readCSV_files_3 (path):   #Filtered signals
## Avarage Sample- Read 
    csv_files = glob.glob(os.path.join(path, "*.csv"))
    AS_45=[] 
    AS_5=[] 
    AS_4=[] 
    ALARM=[] 
    ALARMT=[]
    RATIO=[]
    RATIO4=[]
    RATIO5=[]

    ALRM_COND_REG=[] # 15
    ALRM_COND_CLN=[] # 14
    FIRE_INDICATION=[] # 13
    PULSE_COUNT=[] # 12
    PULSE_ALARM=[] # 11
    ALARM_STATE=[] # 10
    UFAST_BLOCK=[] # 9
    UFAST_ALARM=[] # 8
    STARTUP=[] # 7
    BIT=[] # 6
    INTERFERENCE=[] # 5
    INTERFERENCE_COND=[] # 4
    PRE_ALARM=[] # 1
    TEMPR_SLEW_RATE=[] # 0
    EMI_P=[]
    EMI_C=[]
    EMI_A=[]
    SUS=[]

    NON_A=[]
    ALARM_A=[]


    h=0 
    index_alarm=None
    # loop over the list of csv files
    for f in csv_files:
        
        
        
    
#         # read the csv file
        df = pd.read_csv(f)
        df=df.fillna(0)
#         t=(df).to_numpy()
#         if (t[0,1])==1:
#             index_45=6
#             index_4=7
#             index_5=8
#             index_alarm=5
#             index_ratio=12
#         elif (t[0,2])==1:
#             index_45=7
#             index_4=8
#             index_5=9
#             index_alarm=6
#             index_ratio=13
#         elif t[0,0]=='c':
#             index_45=8
#             index_4=10
#             index_5=12
#             index_alarm=1
#             index_ratio=13
            
#         else:
#             index_45=5
#             index_4=6
#             index_5=7
#             if (type(t[0,1])==float):
#                 index_alarm=4
#             else:
#                 index_alarm=1
#             index_ratio=11


        if ("FILTERED_SAMPLE_5") in df.keys():
            fs_45="FILTERED_SAMPLE_4.5"
            fs_4="FILTERED_SAMPLE_4"
            fs_5="FILTERED_SAMPLE_5"
            alarmi="ALARM"
            rat45="RATIOS_4.5"
            rat4="RATIOS_4"
            rat5="RATIOS_5"
        if ("filtSignal[S5]") in df.keys():
            fs_45="filtSignal[S45]"
            fs_4="filtSignal[S4]"
            fs_5="filtSignal[S5]"
            alarmi="Detection (Alarm) flags"
            rat45="ratio45to4"
            rat4="ratio45to5"
            rat5="ratio4to5"
            
        if ("EMI_Prob") in df.keys():
            emi_p="EMI_Prob"
            emi_c="EMI_Count"
        
            emi_a=np.zeros((df["alarmExtFlags"]).to_numpy().size)
            sus=np.zeros((df["alarmExtFlags"]).to_numpy().size)
            
            
            for i in range((df["alarmExtFlags"]).to_numpy().size):
                bit=np.binary_repr(int(((df["alarmExtFlags"]).to_numpy())[i]),width=16)
                bit2=np.array(list(bit))

                if (int(bit2[13])==1):
                    emi_a[i]=1
                    
                if (int(bit2[14])==1):
                    sus[i]=1
            
            
            EMI_P.append((df[emi_p]).to_numpy())
            EMI_C.append((df[emi_c]).to_numpy())
            EMI_A.append([emi_a])
            SUS.append([sus])
        else:
            EMI_P.append(np.zeros((df[alarmi]).to_numpy().size))
            EMI_C.append(np.zeros((df[alarmi]).to_numpy().size))
            EMI_A.append(np.zeros((df[alarmi]).to_numpy().size))
            SUS.append(np.zeros((df[alarmi]).to_numpy().size))
    
            
        
        AS_45.append((df[fs_45]).to_numpy())
        AS_4.append((df[fs_4]).to_numpy())
        AS_5.append((df[fs_5]).to_numpy())
        
        ALARM.append((df[alarmi]).to_numpy())
        RATIO.append((df[rat45]).to_numpy())
        RATIO4.append((df[rat4]).to_numpy())
        RATIO5.append((df[rat5]).to_numpy())
        
        alarm=np.zeros((df[alarmi]).to_numpy().size)
        alarma=np.zeros((df[alarmi]).to_numpy().size)
        nona=np.zeros((df[alarmi]).to_numpy().size)
        alrm_cond_reg=np.zeros((df[alarmi]).to_numpy().size)
        alrm_cond_cln=np.zeros((df[alarmi]).to_numpy().size)
        bitt=np.zeros((df[alarmi]).to_numpy().size)
        fire_indication=np.zeros((df[alarmi]).to_numpy().size)

        for i in range((df[alarmi]).to_numpy().size):
            bit=np.binary_repr(int(((df[alarmi]).to_numpy())[i]),width=16)
            bit2=np.array(list(bit))

            if (int(bit2[10])==1):
                alarm[i]=1
                nona[i]=0
                alarma[i]=0
            
            if (int(bit2[15])==1):
                alrm_cond_reg[i]=1
            if (int(bit2[14])==1):
                alrm_cond_cln[i]=1
            if (int(bit2[6])==1):
                bitt[i]=1
            if (int(bit2[13])==1):
                fire_indication[i]=1

        ALRM_COND_REG.append(alrm_cond_reg)   
        ALRM_COND_CLN.append(alrm_cond_cln)
        BIT.append(bitt)
        ALARMT.append(alarm)
        FIRE_INDICATION.append(fire_indication)
        NON_A.append(nona)
        ALARM_A.append(alarma)
        #h=h+1
        #w=t[:,3].size
        
    return AS_45, AS_4, AS_5,ALARMT,FIRE_INDICATION, NON_A, ALARM_A, EMI_P, EMI_C, EMI_A, SUS
```
